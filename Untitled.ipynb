```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score

# Load Data
expense_df = pd.read_csv("data/expenses.csv")
incidents_df = pd.read_csv("data/incidents.csv")

# Standardize column names (convert to lowercase and replace spaces with underscores)
incidents_df.columns = incidents_df.columns.str.strip().str.lower().str.replace(" ", "_")
expense_df.columns = expense_df.columns.str.strip().str.lower().str.replace(" ", "_")

# Merge datasets on incident_id
df = pd.merge(incidents_df, expense_df, on="incident_id", how="left")

# Rename columns for clarity
df.rename(columns={"season_x": "season", "weather_x": "weather"}, inplace=True)

# Handle missing values
df["amount"].fillna(0, inplace=True)
df["category"].fillna("Unknown", inplace=True)

# Convert `hospital_transport` to binary (Yes = 1, No = 0)
df["hospital_transport"] = df["hospital_transport"].apply(lambda x: 1 if x == "Yes" else 0)

# Convert `injury_severity` (ordinal 1-10, already numeric)
df["injury_severity"] = pd.to_numeric(df["injury_severity"], errors="coerce").fillna(0)

# Convert categorical columns (season, weather, injuries)
df["season"] = df["season"].apply(lambda x: 1 if x == "Winter" else 0)  # 0 for Summer, 1 for Winter

# One-hot encode weather and injuries
weather_encoder = OneHotEncoder(drop='first', dtype=int)  # Avoid multicollinearity by dropping the first category
weather_encoded = weather_encoder.fit_transform(df[['weather']]).toarray()
weather_columns = weather_encoder.get_feature_names_out(input_features=["weather"])

injury_encoder = OneHotEncoder(drop='first', dtype=int)
injury_encoded = injury_encoder.fit_transform(df[['injuries']]).toarray()
injury_columns = injury_encoder.get_feature_names_out(input_features=["injuries"])

# Concatenate the one-hot encoded columns to the dataframe
df = df.join(pd.DataFrame(weather_encoded, columns=weather_columns))
df = df.join(pd.DataFrame(injury_encoded, columns=injury_columns))

# Select relevant columns (features) for the model
# Define the 'staff_needed' variable
# Ensure columns are numeric
df["injuries"] = pd.to_numeric(df["injuries"], errors="coerce").fillna(0)
df["injury_severity"] = pd.to_numeric(df["injury_severity"], errors="coerce").fillna(0)
df["response_time"] = pd.to_numeric(df["response_time"], errors="coerce").fillna(0)
df["hospital_transport"] = pd.to_numeric(df["hospital_transport"], errors="coerce").fillna(0)
df["hospital_stay_length"] = pd.to_numeric(df["hospital_stay_length"], errors="coerce").fillna(0)

# Define the 'staff_needed' variable
df["staff_needed"] = (
    df["injuries"] * 0.4 + 
    df["injury_severity"] * 0.3 +
    df["response_time"] * 0.2 +
    df["hospital_transport"] * 0.5 +
    df["hospital_stay_length"] * 0.1
).astype(int)  # Convert to integer for staffing levels

# Now proceed with model training as before
features = ["injury_severity", "response_time", "hospital_transport", "hospital_stay_length", "amount"] + list(weather_columns) + list(injury_columns)
X = df[features]
y = df["staff_needed"]  # Now "staff_needed" exists

# Normalize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split Data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train a Predictive Model (Random Forest Regressor)
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Model Evaluation
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {mae:.2f}")
print(f"R² Score: {r2:.2f}")

# Now proceed with model training as before
features = ["injury_severity", "response_time", "hospital_transport", "hospital_stay_length", "amount"] + list(weather_columns) + list(injury_columns)
X = df[features]
y = df["staff_needed"]  # Now "staff_needed" exists

# Normalize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split Data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train a Predictive Model (Random Forest Regressor)
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Model Evaluation
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {mae:.2f}")
print(f"R² Score: {r2:.2f}")

features = ["injury_severity", "response_time", "hospital_transport", "hospital_stay_length", "amount"] + list(weather_columns) + list(injury_columns)
X = df[features]
y = df["staff_needed"]  # Assuming "staff_needed" is the target variable

# Normalize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split Data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Train a Predictive Model (Random Forest Regressor)
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {mae:.2f}")
print(f"R² Score: {r2:.2f}")




```

```python
import matplotlib.pyplot as plt

# Data for current ticket sales (in dollars)
current_sales = {
    "Climbing": {"Summer": 320, "Winter": 2112.5},
    "Hiking": {"Summer": 280, "Winter": 1302.5},
    "Mountain Biking": {"Summer": 250, "Winter": 2471.67},
    "Skiing": {"Summer": 475, "Winter": 2247.5},
    "Snowboarding": {"Summer": 748.57, "Winter": 3633.75},
}

# Define the growth percentage (e.g., 10%)
growth_percentage = 0.10

# Calculate the new ticket sales
new_sales = {}
for activity, seasons in current_sales.items():
    new_sales[activity] = {}
    for season, sale in seasons.items():
        new_sales[activity][season] = sale + (sale * growth_percentage)

# Prepare data for visualization
activities = list(current_sales.keys())
seasons = ["Summer", "Winter"]

current_sales_list = [current_sales[activity][season] for activity in activities for season in seasons]
new_sales_list = [new_sales[activity][season] for activity in activities for season in seasons]

# Create a bar chart
fig, ax = plt.subplots(figsize=(10, 6))
width = 0.35  # Bar width

# Positions for the bars on the x-axis
x = range(len(activities) * len(seasons))

# Plot current and new sales
ax.bar(x, current_sales_list, width, label='Current Sales', color='blue')
ax.bar([i + width for i in x], new_sales_list, width, label='New Sales (10% Increase)', color='green')

# Labeling the chart
ax.set_xlabel('Activity & Season')
ax.set_ylabel('Ticket Sales ($)')
ax.set_title('Ticket Sales: Current vs. 10% Increase by Activity & Season')
ax.set_xticks([i + width / 2 for i in range(0, len(x), 2)])
ax.set_xticklabels([f'{activity} - {season}' for activity in activities for season in seasons], rotation=45, ha="right")
ax.legend()

# Display the chart
plt.tight_layout()
plt.show()
```
